**Development Process**

The data life cycle will be the development process in use for this assessment. It consists of 7 main steps beginning with creating an understanding of the businesses needs to putting in the work gathering data, cleaning data, analysing data etc. Then, presenting findings back to the business usually in the form of graphs. For version control i will be using Git to create a repository to which i can stage and commit changes to. Because of colabs collaboration with GitHub i will be able to track changes and revert to previous versions incase of mistakes.

**Data Life Cycle**

*   Business Understanding - Conducting interviews with the business to develop an understanding of the problem and create objectives and goals that align with stakeholder needs.

*   Data Mining - Indetifying sources of data to later gather and scrape, then conducting a quality assessment of the data to eventually transform the data and explore the data relationships.

*   Data Cleaning - Fix the data by correcting missing values, duplicate values, redundant data, inconsistent data, incorrect data. Eventually standardising and normalising the data.

*   Data Exploration - Analysing the collected and corrected data visually, identifying patterns and trends, to form hypothesis about solutions to the given problem.

*   Feature Engineering - Creating new and improved features using your hypothesied solution and the corrolating raw data.

*   Predictive Modelling - Train an appropriate machine learning model by splitting data into training and testing sets, evaluating how the model performed and tweaking the neccessary aspects to finalise the model.

*   Data Visualisation - Simplify the data and place it in an appropriate graph to help explain and visualise the solution backed up with expected results to the business.
